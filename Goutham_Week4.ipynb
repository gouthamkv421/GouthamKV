```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the excel file
df = pd.read_excel("C:\\Users\\abcd\\Downloads\\Telco-Customer-Churn.xlsx")
```

```python
# Displaying the datset
print(df)
```

```python
df["Churn"].value_counts()
```

```python
# Comparing the Churn count w.r.t.gender
plt.figure(figsize=(6,3))
sns.countplot("gender", data=df, hue = 'Churn')
plt.title("Plot of Churn Vs Gender",fontsize=15)
```

```python
df["SeniorCitizen"].value_counts()
```

```python
# No.of female senior citizen in the dataset
df_female_senior=sns.countplot(data=df,x="SeniorCitizen",hue="gender")
plt.title("Plot of Senior Citizen Vs Gender",fontsize=15)
for i in df_female_senior.containers:
    df_female_senior.bar_label(i)
plt.show()
```

```python
# No of female senior citizen = 568
```

```python
plt.figure(figsize=(10,8))
plt.title("Comparison of Tenure with Total Charges",fontsize=16)
plt.hist(df["tenure"],color='blue',rwidth=0.7,cumulative=True)
plt.xlabel("Tenure",fontsize=12) 
plt.ylabel("TotalCharges",fontsize=12)
plt.show()
```

```python
# Total charge is directly propotional tenure.
```

```python
df1=df.loc[df['SeniorCitizen'] == 1.0] 
df1["Contract"].value_counts()
```

```python
# Month-to-month contract is prefered by Senior citizen
```

```python
# Analysis of Payment method prefered
plt.figure(figsize=(8,6))
plt.xticks(rotation=20)
df_payment_method=sns.countplot(df["PaymentMethod"])
plt.title("Comparison of different Payment Methods",fontsize=15)
for i in df_payment_method.containers:
    df_payment_method.bar_label(i)
```

```python
# It is evident that, electronic check method is prefered by most of them. Other payment options have equal acceptance.
```
