```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# To read the file
data = pd.read_csv("C:\\Users\\abcd\\Downloads\\StudentsPerformance.csv")
data
```

```python
# No. of Male
data1 = data['gender'].value_counts()['male']
data1
```

```python
# No. of Female
data2 = data['gender'].value_counts()['female']
data2
```

```python
# Parental level of Education
data3 = data['parental level of education'].value_counts()
data3
```

```python
# Scoring variation  on Math/Reading/ Writing based on gender
data4 = data.groupby('gender')['math score','reading score','writing score'].std()
data4
```

```python
# Scoring variation on Math/Reading/ Writing based on test preparation
data5 = data.groupby('test preparation course')['math score','reading score','writing score'].std()
data5
```

```python
# Scores average for Maths/Reading/Writing based on gender
data6 = data.groupby('gender')['math score','reading score','writing score'].mean()
data6
```

```python
# Scores average for Maths/Reading/Writing based on Test preparation course
data7 = data.groupby('test preparation course')['math score','reading score','writing score',].mean()
data7
```

```python
# List of 25% students who score highest marks in Maths
data8 = data.nlargest(250,['math score'])
data8
```
